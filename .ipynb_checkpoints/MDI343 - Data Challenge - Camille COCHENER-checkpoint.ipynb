### Summary of the analysis

My analyses took place in several stages.

***--- Data exploration***

First, I explored the data. I checked that there were no <b>missing values</b>. If there were, an imputation technique for missing values should have been implemented because the algorithms do not handle the presence of missing values well.  
Next, since one line represented the comparison between two images, and we had groups of variables specific to each of the two images, it seemed interesting to me to <b>assign identifiers to the images</b> in order to understand the structure of the dataset and to check whether some images in the training dataset were common to the images in the test dataset. In the end, it turned out that very few images were common to both datasets. Only the different quality and match scores would be useful in learning the model.   
Then, the different variables in the dataset were analysed in a descriptive way. The <b>response variable</b>, a binary variable indicating whether the two images match or not, was very <b>unbalanced</b> (about 6% of the images match). It was therefore relevant to choose an appropriate metric for this case, such as TPR (also called Recall or Sensitivity).  
In addition, each <b>explanatory variable</b> was analyzed, again in relation to the response variable, in order to determine which ones could characterize certain classes. It appeared that quality scores did not seem to play an obvious role in the discrimination of the two classes, but that matching scores clearly had a different distribution depending on whether or not the images matched.   
Finally, <b>correlation analysis</b> showed that quality scores were not correlated with each other. On the other hand, the match scores were highly correlated with each other when the images were matched. The correlation structure of the matching scores depended on the class. 
    
In conclusion, little preparation or data transformation was required. 
Once I understood the dataset, I was able to start building a classifier.

**--- Modeling**

In the case of unbalanced data, <b>ensemble methods</b> can often give good results because they avoid overfitting and solve often complicated problems. <br>
In this category of algorithms, I have chosen boosted based techniques because they are simple to use, provide models that are often very generalizable and are not subject to overfitting. These methods are not very sensitive to noise and outliers. <br>
I chose to use two powerful algorithms: the <b>XGBoost</b> and the <b>CatBoost algorithm</b>. The best performance was obtained with XGBoost.
 
  
***What is XGBoost and CatBoost?***
  
XGBoost is a scalable machine learning system for tree boosting. It has been recognized in a number of machine learning and data mining challenges, on a wide range of problems. The most important factor behind the success of XGBoost is its scalability in all scenarios due to several important systems and algorithm optimizations from gradient boosting. It is an ensemble of decision trees algorithm where new trees fix errors of those trees that are already part of the model. Trees are added until no further improvements can be made to the model.<br>
All the mathematical details is in this <a href="https://www.kdd.org/kdd2016/papers/files/rfp0697-chenAemb.pdf">paper</a>.   
  
But more recently (2018), an other algorithm called CatBoost (an open-source library) got a lot of attention. As all existing implementations of gradient boosting face a statistical issue, a <b>prediction shift</b> of the learned model caused by a special kind of target leakage, CatBoost brings a solution to this problem, the <b>ordered boosting</b>, which is a modification of the standard gradient boosting algorithm avoiding target leakage. It also allows to process categorical features.    
This algorithm <b>outperforms the existing state-of-the-art implementations of gradient boosted decision trees</b> like XGBoost and LightGBM on multiple machine learning tasks. <br>
For mathematical details, please refer to the following paper <a href="https://arxiv.org/abs/1706.09516">here</a>.

  
***XGBoost training***
  
For training, I chose to split the Xtrain dataset into two parts: a training dataset and a validation dataset. <br>
In order to avoid biasing the generalization error, I used a cross-validation. Nevertheless, the results on the different folds were very stable (more or less $10^{-4}$). Considering the time taken by the cross-validation, I considered that a simple cut was sufficient and gave a good estimate of the generalization error. <br>
I have been careful to <b>keep the proportions of each class</b> in each of the training and validation parts.
  
***Hyperparameters choice***
  
Building a grid search being too computationally intensive on this dataset, I set the model parameters by hand. First, I left the number of iterations at 1000 and the learning rate at 0.1, then I varied the main parameters such as the depth of the tree, the regularization parameter, the number of sample and features to build a tree. Once an optimum was found, I started to decrease the learning rate and to increase the number of iterations in order to obtain the best possible quality.
<br>

***Evaluation***

The evaluation metric is the best true positive rate when the false positive rate is lower than $10^{-4}$. 
With my algorithm, I got a TPR of 0.9873114924596984 on the validation set, and a TPR of 0.878006370505 on the test set.   
The difference between the two scores is 0.10762771. I suppose this is due to the fact that part of the distributions of the positive and negative classes in the test set is not captured or is not present in the training set.

#### 4.2 Model training

##### <b><i>Catboost</b>

```python
# ---- Best hyperparameters
# iterations=2000
# learning_rate=0.4
# depth=12
# l2_leaf_reg=2
# bagging_temperature=0
# random_strength=4
# loss_function='Logloss'
# task_type='GPU'
# random_seed=0
```

```python
# Define the model
model_cat = CatBoostClassifier(
    iterations=5000, # Number of iterations
    learning_rate=0.4, # Learning rate 
    depth=12, # Depth of the trees
    l2_leaf_reg=2, # Regularization parameter
    bagging_temperature=0,
    random_strength=4,
    loss_function='Logloss', # The loss function
    task_type='GPU',
    random_seed=0,
    early_stopping_rounds=40,
    use_best_model=True
)
```

```python
# Train the model on the validation set
model_cat.fit(
      train_data, 
      train_labels, 
      verbose=True, 
      eval_set=(valid_data, valid_labels)
)
```

```python
# Predictions on the validation set
y_pred = model_cat.predict_proba(valid_data)[:, model_cat.classes_ == 1][:, 0]
print(y_pred)
```

```python
# Compute the performance metric
ypred_scoreordered = valid_labels[np.argsort(y_pred)]
N = np.sum(valid_labels == 0)
P = np.sum(valid_labels == 1)
FP = 0
TP = 0
for i in range(len(y_pred) - 1, -1, -1):
    if (ypred_scoreordered[i] == 1):
        TP = TP + 1
    else:
        FP = FP + 1
    if (FP / N > 10**-4):
        FP = FP - 1
        break
        
print("CATBOOST RESULTS ----")
print("Number of true positive : {}".format(TP))
print("Number of false positive : {}".format(FP))
print("Number of total positive : {}".format(P))
print("Number of total negative : {}".format(N))
print("For the smallest FPR <= 10^-4 (i.e., ", FP / N, ") TPR = ", TP / P, ".", sep = "")
```
